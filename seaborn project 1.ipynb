```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("accident.csv")
df
```

### Isna,Dtypes,Describe,columns

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df.describe
```

```python
df.columns
```

### Fillna

```python
df['Gender'].fillna('Female',inplace=True)
```

```python
df['Speed_of_Impact'].fillna(df['Speed_of_Impact'].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
df['Speed_of_Impact']=df['Speed_of_Impact'].astype(int)
```

```python
df.dtypes
```

## Seaborn Graphs

### Histogram

```python
sns.histplot(data=df,x='Speed_of_Impact',stat='frequency',kde=True)
plt.show()
```

### Scatter Plot

```python
sns.scatterplot(data=df,x='Age',y='Speed_of_Impact',hue='Gender')
plt.show()
```

### Barplot

```python
sns.barplot(x='Gender',y='Speed_of_Impact',data=df,hue='Helmet_Used',color='skyblue',saturation=0.9)
plt.xlabel("Gender")
plt.ylabel("speed of impact")
plt.title("Relationship between Gender and speed of impact ")
plt.show()
```

```python
sns.barplot(x='Helmet_Used',y='Age',data=df,hue='Gender',ci=None,palette='muted',saturation=0.9)
plt.xlabel("Helmet_used")
plt.ylabel("Age")
plt.title("Relationship between Age and Helmet Used ")
plt.show()
```

```python
sns.barplot(x='Helmet_Used',y='Speed_of_Impact',data=df,hue='Gender',color='lightgreen',saturation=0.9)
plt.xlabel("Helmet_used")
plt.ylabel("Speed of impact")
plt.title("Relationship between Speed of impact and Helmet Used ")
plt.show()
```

### Countplot

```python
sns.countplot(x='Gender',data=df)
plt.show()
```

```python
sns.countplot(x='Helmet_Used',data=df,facecolor='hotpink',edgecolor='black')
plt.show()
```

```python
sns.countplot(x='Seatbelt_Used',data=df,facecolor='red',edgecolor='black')
plt.show()
```

```python
sns.countplot(x='Survived',data=df,facecolor='lightgreen',edgecolor='black')
plt.show()
```

### Swarmplot and Stripplot

```python
sns.swarmplot(x='Gender',y='Speed_of_Impact',data=df)
plt.show()
```

```python
sns.swarmplot(x='Seatbelt_Used',y='Speed_of_Impact',data=df,hue='Gender')
plt.show()
```

```python
sns.stripplot(x='Survived',y='Speed_of_Impact',data=df,hue='Gender')
plt.show()
```

### Boxplot and Violinplot

```python
sns.boxplot(x='Gender',y='Speed_of_Impact',hue='Gender',showmeans=True,data=df)
```

```python
sns.violinplot(x='Gender',y='Speed_of_Impact',hue='Gender',data=df)
```

### KDEplot and Regression plot

```python
sns.kdeplot(data=df,x='Age',y='Speed_of_Impact',shade=True,color='skyblue')
plt.show()
```

```python
sns.regplot(data=df,x='Age',y='Speed_of_Impact')
plt.show()
```

### Pairplot and Jointplot

```python
sns.pairplot(df)
```

```python
sns.jointplot(data=df,x='Age',y='Speed_of_Impact',kind='scatter')
```
